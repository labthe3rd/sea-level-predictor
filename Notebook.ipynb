#### Import dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

#### Retrieve data from .csv file

```python
df = pd.read_csv("epa-sea-level.csv")
df
```

#### Create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the Y axis

```python
ax1 = df.plot(kind='scatter',x='Year',y='CSIRO Adjusted Sea Level',c='DarkBlue')
```

#### Use the lingress function from scipy.stats to get the slope and y-intercept of the line of best fit, then plot it over the scatter plot.

```python
res = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
res

plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], c='DarkBlue', label='Data Points')

# Plotting the regression line
x = np.linspace(min(df['Year']), max(df['Year']), 100)  # Generating a range of x values
y = res.intercept + res.slope * x                       # Calculating corresponding y values using the regression equation
plt.plot(x, y, 'r', label='Regression Line')

# Annotating the plot
plt.title("Year vs CSIRO Adjusted Sea Level with Regression Line")
plt.xlabel("Year")
plt.ylabel("CSIRO Adjusted Sea Level")
plt.legend()

plt.show()
```
